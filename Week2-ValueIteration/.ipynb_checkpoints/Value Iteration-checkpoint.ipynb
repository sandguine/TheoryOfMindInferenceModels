```python
actionSpace = list(set(itertools.permutations([-1,-1, 0, 0, 1, 1], 2)))
stateSpace = list(itertools.product(range(3), range(3)))

valueTable = {coord:0 for coord in stateSpace}
qTable = {coord: {act:0 for act in actionSpace} for coord in itertools.product(range(3), range(3))}
reward = {coord: {act:getCostOfAction(coord, (coord[0]+act[0], coord[1]+act[1]), (2,2)) for act in actionSpace} for coord in itertools.product(range(3), range(3))}
#reward = {coord: -1 for coord in stateSpace}
#reward[(1,2)] = 0
```

```python
def getCostOfAction(currentState, nextState, goal):
    if nextState == goal:
        return(0)
    distance = sum([abs(currentState[i] - nextState[i]) for i in range(len(currentState))])
    if distance == 0:
        cost = -1
    else:
        cost = -1/(distance)**.5
    return(cost)
```

```python
class setupRewardTable(object):
    def __init__(self, gridWidth, gridHeight, actionSet, goalState):
        self.gridWidth = gridWidth
        self.gridHeight = gridHeight
        self.stateSet = list(itertools.product(range(self.gridWidth), range(self.gridHeight)))
        self.actionSet = actionSet
        self.goalState = goalState
        
    def __call__(self):
        rewardTable = {state: self.getStateReward(state) for state in self.stateSet}
        return(rewardTable)
    
    def getNextState(self, state, action):
        if validAction(state, action, self.gridWidth, self.gridHeight):
            return(elementwiseTupleAddition(state,action))
        return(state)  

    def getStateReward(self, state):
        actionRewardTable = {action: self.applyRewardFunction(state, action) for action in self.actionSet}
        return(actionRewardTable)

    def applyRewardFunction(self, currentState, action):
        nextState = self.getNextState(currentState, action)
        
        if nextState == self.goalState:
            return 0
        else:
            distance = sum([abs(currentState[i] - nextState[i]) for i in range(len(currentState))])
            if distance == 0:
                reward = -1
            else:
                reward = -1/(distance)**.5
            return(reward)


def validAction(state, action, gridWidth, gridHeight):
    xCoordinate, yCoordinate = elementwiseTupleAddition(state, action)
    if(xCoordinate < 0 or  xCoordinate >= gridWidth):
        return(False)
    if(yCoordinate < 0 or yCoordinate >= gridHeight):
        return(False)
    return(True)

def elementwiseTupleAddition(firstTuple, secondTuple):
    assert len(firstTuple) == len(secondTuple),"Tuples are of different length."
    return(tuple([firstTuple[i] + secondTuple[i] for i in range(len(firstTuple))]))
```

```python
gridWidth = 2
gridHeight = 2
allActions = list(set(itertools.permutations([-1,-1, 0, 0, 1, 1], 2)))
goalState = (0,1)

getReward = setupRewardTable(gridWidth, gridHeight, allActions, goalState)
reward = getReward()
```

```python
class setupTransition(object):
    def __init__(self, gridWidth, gridHeight, actionSet):
        self.gridWidth = gridWidth
        self.gridHeight = gridHeight
        self.stateSet = list(itertools.product(range(self.gridWidth), range(self.gridHeight)))
        self.actionSet = actionSet

    def __call__(self):
        transitionTable = {state: self.getStateTransition(state) for state in self.stateSet}
        return(transitionTable)
    
    def getNextState(self, state, action):
        if validAction(state, action, self.gridWidth, self.gridHeight):
            return(elementwiseTupleAddition(state,action))
        return(state)  

    def getStateTransition(self, state):
        actionTransitionDistribution = {action: self.getStateActionTransition(state, action) for action in self.actionSet}
        return(actionTransitionDistribution)

    def getStateActionTransition(self, currentState, action):
        allNextStates = [elementwiseTupleAddition(currentState, a) for a in self.actionSet]
        nextState = self.getNextState(currentState, action)
        
        stateActionTransitionDistribution = {state: 0 for state in allNextStates}
        stateActionTransitionDistribution[nextState] = 1
        
        return(stateActionTransitionDistribution)

def validAction(state, action, gridWidth, gridHeight):
    xCoordinate, yCoordinate = elementwiseTupleAddition(state, action)
    if(xCoordinate < 0 or  xCoordinate >= gridWidth):
        return(False)
    if(yCoordinate < 0 or yCoordinate >= gridHeight):
        return(False)
    return(True)

def elementwiseTupleAddition(firstTuple, secondTuple):
    assert len(firstTuple) == len(secondTuple),"Tuples are of different length."
    return(tuple([firstTuple[i] + secondTuple[i] for i in range(len(firstTuple))]))
```

```python
gridWidth = 2
gridHeight = 2
allActions = list(set(itertools.permutations([-1,-1, 0, 0, 1, 1], 2)))

getTransition = setupTransition(gridWidth, gridHeight, allActions)
transition = getTransition()
```
