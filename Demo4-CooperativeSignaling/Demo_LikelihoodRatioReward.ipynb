# Signaling Policies
## Introduction
In the goal inference module, a third party infers another agent's goal as an outsider by observing the trajectory of states they passed across time. Here, the problem is approached from a first person point of view  -- modeling the actions an agent wants to take. However, the agent we are modeling is a cooperative one, trying to help an observer understand its goal. This means our chosen actions should try to signal our true goal while ruling out other goals as quickly and unambiguously as possible. 

An agent's policy should change to reflect the reduction of ambiguity between goals. However, we do not directly control the policy. Instead, we perform value iteration to find the relative values of states and actions and use these to derive our policy. The environment dynamics and reward function define what makes a state valuable. However, an agent is not in control of the dynamics of the environment, so we change the reward function to reflect our desire to send cooperative signals.

## Tailoring the Reward Function for Signaling: Likelihood Ratio
Recall that reward is defined: $R( s^\prime, a, s)$, so we have a reward for each state, action, next state set. Before, the components of this reward function were the large reward or cost of special states (goal or trap) as well as the small cost of moving. Now, we want to add another term to our reward: a measure of how much ambiguity we are reducing by taking a particular action. When we performed goal inference, we found $P(s_{t+1}|s_t, g)$, the probability of observing the $s_{t+1}$ given the current state $s_t$ and the goal $g$. 

\begin{equation}\label{probNextState}
    P(s_{t+1}|s_t, g) = \sum_{a_t \in A_{s_t}}P(s_{t+1}|s_t, a_t) \pi(a_t|s_t, g)
\end{equation}

There we used it to evaluate with a known $s_{t+1}$, which goal was most likely. Now we can take the ratio of:

\begin{equation}\label{ratio}
    r_{info} = \frac{P(s_{t+1}|s_t, g_{true})}{\sum_{g \in G}P(s_{t+1}|s_t, g)}
\end{equation}

This is the likelihood ratio of the next state given the actual goal compared to all goals. Thus, it should be more rewarding to try to get to next states that are likely for the true goal and unlikely for the other goals.

For each state-action-next state combination, we take the given reward and update it to reflect this additional $r_{info}$.
\begin{equation}\label{reward}
    r_{new}=r_{original}+\alpha r_{info}
\end{equation}

Where $\alpha$ affects how much preference is given to the information added. From this, you can create a new reward function for each state-action-next state, and using that new reward function, create a new policy that takes intention to signal into account.

# Examples
## Original (non-signaling) Policies
These policies are generated in the same manner as those used in the goal inference module (before posterior estimation). The environment space is the same as environment 2 (with a gap) in the goal inference module. The only difference is that the action space is reduced to cardinal directions and the no movement action (whereas previously we included diagonal movements). The reduced action space allows us to more easily visualize the difference between policies before and after the change in reward

```python
gridWidth = 7
gridHeight = 6
allActions = [(1,0), (0,1), (-1,0), (0,-1), (0,0)]
trapStates = [(3,0), (3,1), (3,3)]
otherGoals = [(6,1), (6,4), (1,5)]
stateSet = list(itertools.product(range(gridWidth), range(gridHeight)))

convergenceThreshold = 10e-7
gamma = .9
beta = 2

getTransition = SetupDeterministicTransitionByGrid(gridWidth, gridHeight, allActions)
transition = getTransition()
```

## Goal A
### Policy Setup

```python
goalStateA = (6,4)
getRewardA = SetupRewardTableWithDistanceMoveCost(transition, allActions, [goalStateA],trapStates)
rewardA = getRewardA()
valueA = {state:0 for state in rewardA.keys()}

performValueIterationA = BoltzmannValueIteration(transition, rewardA, valueA, convergenceThreshold, gamma, beta)
optimalValuesA, policyA = performValueIterationA()
```

### Visualizations

Value table and policy

```python
visualizeValueTableMultipleGoals(gridWidth = gridWidth, gridHeight = gridHeight, goalState = goalStateA, otherGoals = otherGoals, trapStates = trapStates, valueTable = optimalValuesA)
visualizePolicy(stateSet, policyA, goalStateA, otherGoals, trapStates)
```
