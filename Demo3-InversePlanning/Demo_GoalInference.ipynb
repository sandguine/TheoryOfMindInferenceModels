# Goal Inference

## Introduction
This demo is a model of Goal Inference from a third person perspective. You are an observer who sees an agent travel through a series of states and must reason about which state they are likely trying to reach. This environment contains a set of possible goals and trap (costly) states that act as barriers. 

This is an implementation of Model 1 of a 2007 paper by Baker, Saxe, and Tenenbaum:  [Action Understanding as Inverse Planning](https://www.sciencedirect.com/science/article/pii/S0010027709001607).

## Goal Inference as Posteriors of Bayesian Inference
On the highest level, this is a Bayesian Inference of the form:
\begin{equation}\label{Bayes:Goal}
    P(\textrm{Goal}|\textrm{Actions, Environment}) \propto P(\textrm{Actions}|\textrm{Goal, Environment})P(\textrm{Goal}|\textrm{Environment})
\end{equation}

## Temporal Effects
However, we do not observe the sequence of actions an agent takes. Instead, we observe the trajectory of their states. Thus, we can rewrite our Bayesian Inference problem in terms of the state sequence:

\begin{equation}\label{prior}
    P(g|S_{1:T})  \propto P(S_{2:T}|s_1, g)P(g)
\end{equation}

Here, $s_1$ is the initial state and $S_{1:T}$ is a sequence of observed states. We do not need to explicitly worry about the environment because it is incorporated into the reward and transition (e.g. which states have high costs, which states are goal states, and the state space itself). This is now a sequential Bayesian updating problem. Like typical Bayesian Inference, it is broken down into posterior $\propto$ (Likelihood)(Prior), however, this is done at each time point. 

Given an initial state and a prior over the goals, we can sequentially update the probability of the goal given this information. As a result $P(S_{2:T}|s_1, g)$ breaks into:

\begin{equation}\label{sequentialLikelihood}
    P(S_{2:T}|s_1, g) = \prod_{t=1}^{T-1}P(s_{t+1}|s_t, g)
\end{equation}

Where:
\begin{equation}
    P(s_{t+1}|s_t, g) = \sum_{a_t \in A_{s_t}}P(s_{t+1}|s_t, a_t) \pi(a_t|s_t, g)
\end{equation}

In the equation immediately above, the first term is the transition and the second is the policy. To find the policy, we perform value iteration until we get convergence of the state values. 

## Using a Softmax Policy
Instead of a deterministic optimal policy, we assume agents have a probability distribution over actions based on the value of taking the action from their current state. This is necessary to infer goals from non-optimal state trajectories. We choose an action according to a Boltzmann policy:

\begin{equation}\label{policy}
    \pi(a_t|s_t, g) \propto e^{\beta Q^{\pi}_g(s_t, a_t)}
\end{equation}

The Boltzmann Policy draws actions from a softmax and is commonly used to model approximately rational decision making. The $\beta \in [0, \infty]$ parameter is a metric for how rational we believe the actor to be. At $\beta = 0$, the actor is modeled as taking a random walk and at $\beta = \infty$, the agent always takes the expected best action greedily.

# Goal Inference (Inverse Planning)

The goal inference class takes in a transition function describing the dynamics of the environment, a set of  policies corresponding to the possible goals in the environment, a prior over how likely each goal is at time 0, and an actual observed trajectory of actions observed in the environment.

It outputs a posterior distribution over the goals at each time point in the trajectory

## Setup and Helper Functions

```python
def plotPosteriors(posteriors, title="", labels=['A', 'B', 'C']):
    
    x = range(len(posteriors[:, 0]))
    plt.title(title)
    plt.plot(x, posteriors[:,0])
    plt.plot(x, posteriors[:,1])
    plt.plot(x, posteriors[:,2])

    plt.legend(labels)
    plt.show()
```

## Example Trajectory Sampling
Before we used fixed trajectories, instead, we can sample a goal and a corresponding trajectory to solve the same inference problem

```python
def samplePathToGoal(position, policy, transition, goals):
    trajectory = [position]
    while position not in goals:
        #take action probabilisitically
        actions = list(policy[position].keys())
        probOfAction = [policy[position][action] for action in actions]
        actionIndex = np.random.choice(len(actions), size = 1, p = probOfAction)
        sampledAction = actions[int(actionIndex)]
        
        #get new position
        newPosition = list(transition[position][sampledAction].keys())[0]
        
        #update to new position and add to trajectory
        position = newPosition
        trajectory.append(position)
    return(trajectory)
```

Here we can sample a goal and a trajectory for that goal under the specified environment.
